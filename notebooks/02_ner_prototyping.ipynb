# Linguistic Prototyping (Phase B)

This notebook demonstrates how our NLP tools evaluate provisions. We'll use spaCy to tokenize, extract custom entities (NER), and strip Subject-Verb-Object (SVO) mechanics out of regulatory constraints.

```python
import sys
from pathlib import Path

# Ensure modules can be resolved
sys.path.append(str(Path.cwd().parent))

from src.nlp_pipeline.tokenizer import Tokenizer
from src.nlp_pipeline.ner import ConstructionNER
from src.nlp_pipeline.dependency_parser import DependencyParser
```

## 1. Setup Models
Instantiating our core spaCy wrappers and linking the `entity_types.yaml` mapping dictionary.

```python
pattern_path = Path.cwd().parent / "domain" / "entity_types.yaml"

try:
    tokenizer = Tokenizer()
    ner_model = ConstructionNER(patterns_path=str(pattern_path))
    svo_model = DependencyParser()
    print("Models loaded successfully!")
except Exception as e:
    print(f"Warning: Environment might be missing spacy dependencies: {e}")
    print("Run `pip install -e .` from the root directory to fix this.")
```

## 2. Tokenization & Entity Extraction
Testing sentence boundaries and mapping custom parameters.

```python
sample_text = "The minimum width of an exit door shall be 900 mm. Such doors provide a safe exit route."

# 1. Tokenize
try:
    sentences = tokenizer.get_sentences(sample_text)
    print(f"Detected {len(sentences)} Sentences:")
    for s in sentences:
        print(f"  - {s}")

    print("\n" + "="*40 + "\n")

    # 2. NER Extraction
    entities = ner_model.extract_entities(sample_text)
    print("Custom Entities Detected:")
    for e in entities:
        # Print only our specifically engineered labels
        if e['label'] in ["BUILDING_ELEMENT", "MEASUREMENT", "SPATIAL"]:
            print(f"  [{e['label']}] -> {e['text']}")
except Exception as e:
    pass # Environment likely missing ML weights
```

## 3. Dependency Parsing (SVO)
Extracting SVO (Subject-Verb-Object) triplets to power the structural conditions required by Phase D.

```python
try:
    svo_triplets = svo_model.extract_svo(sample_text)

    print("Dependency Mappings:")
    for trip in svo_triplets:
        print(f"  Subject: {trip['subject']}")
        print(f"  Verb:    {trip['verb']}")
        print(f"  Object:  {trip['object']}\n")
except Exception:
    pass
```
